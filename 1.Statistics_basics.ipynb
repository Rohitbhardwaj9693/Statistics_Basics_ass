```python
# Q5. Explain the concept of levels of measurement and give an example of 
# a variable for each level.

# Level of measurment are four type 
# 1. Nominal scale Data
# 2. Odinary scale Data
# 3. Interval scale Data
# 4. Ratio scale Data

# Nominal scale Data: - A nominal scale of data rea categorical data
# * Gendar,Colour
# * order dose not matter in nominal scale of data

# Example 
#         Gender : Male,Female
    
    
# Odinary scale data :- 
# * Ranking is important in odinary scale of data
# * order matter 
# * Differenc cannot be measured in odinary scale of data

# Example of sdinary scale of data 
#     Best,Good,Bad
#     1st,2nd,3rd
    
# Interval scale Data:- 
# * The order matters in interval scale of data
# * Difference can be measured
# * Ratio cannot be measured
# * No true "0 " starting point

# Example
#  Temperature ,variable
#     30f 60:30 = 2:1
#     60f
#     90f 120 - 90  = 30f
#     120f 
    
    
# Ratio scale data
# * The order matter
# * Difference are measured (include ratio )
# * it contain a "0" starting point
# ex 
#  student marks in a class
#     90,50,30,40,50,76,0
#     differencd = 40-30 = 10, 50-30 = 20
#     Ratio = 90/30 = 3:1
    
```
